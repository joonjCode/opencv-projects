```
text2 = pytesseract.image_to_string(img2)
print(text2)
```

```
# Resize the image
img2 = cv2.resize(img2, None, fx=0.7, fy=0.6)
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```
text2 = pytesseract.image_to_string(img2)
print(text2)
```

```
cv2_imshow(gray)
```

Adpative threshold
- useful to get rid of the background

- 85 : pixels
- 11 : noise

```
adaptive_threshold = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,85,12)
cv2_imshow(adaptive_threshold)
```

```
config = '--psm 3'
```

```
text3 = pytesseract.image_to_string(adaptive_threshold, config=config)
print(text3)
```
